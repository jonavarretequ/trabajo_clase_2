```python
#detalles de las columnas de la tabla

columns_elements = soup("td", class_="detail")

#Imprimir los 5 elementos
#columns[:5]

print('Imprimir títulos')
for i in columns_elements[:50]:
    print(i)
```

```python
#Scrapeo complex

# Step 1: Make a GET request
req = requests.get('http://www.ilga.gov/senate/default.asp')

# Step 2: Read the content
src = req.text

# Step 3: Parse the content
soup = BeautifulSoup(src, 'lxml')

# Step 4: Extract column headers
columns_header = soup.find_all("td", class_="header")

# Step 5: Extract column details
columns_elements = soup.find_all("td", class_="detail")

# Step 6: Organize details into rows
rows = []
for i in range(0, 50, 5):  # 50 elements, 5 elements per row
    row = columns_elements[i:i+5]
    rows.append(row)

# Step 7: Print the table
# Print headers
header_titles = [header.get_text(strip=True) for header in columns_header[:5]]
print("\t".join(header_titles))

# Print rows
for row in rows:
    row_data = [element.get_text(strip=True) for element in row]
    print("\t".join(row_data))
```
